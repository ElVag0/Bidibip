```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import TextSplitter 
from typing import Any, List

class CustomTextSplitter(TextSplitter):
    def __init__(self, **kwargs: Any):
        super().__init__(**kwargs)

    def split_text(self, text: str) -> List[str]:
        splits = text.split('\n')
        return splits


docs = TextLoader('../../../saved/generated_save.txt').load()
print("Loaded docs")

sub_docs = CustomTextSplitter(chunk_size = 1494, chunk_overlap  = 20).split_documents(docs)

for page in sub_docs:
    print("##################### PAGE #####################\n", page.page_content)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

persist_directory = "./../../../saved/saved_model/"

vectordb = Chroma.from_documents(documents = sub_docs, embedding = HuggingFaceEmbeddings(), persist_directory = persist_directory)
print("Database generated")

vectordb.persist()
print("Database saved")
```

```python
similar_search_test = vectordb.similarity_search_with_score("Tu aimes les zombies ?", k=10)
print("################# similar search results #################")
for (item, score) in similar_search_test:
    print(score, item.page_content)
print("##########################################################")
```
